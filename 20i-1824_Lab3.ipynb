```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig
from sklearn import datasets
```

# Vector Generation
# Part-1

```python
v1 = np.random.randint(3,7,(1,2))[0]
v1
```

# Orthogonal Vector
# Part-II

```python
v2 = []
v2.append(8)
temp = v2[0]*v1[0]
temp1 = (-1)*temp
v2.append(temp1/v1[1])
v2
```

# Dot Product

```python
ort = np.dot(v1 , v2)
ort
```

# Graphical Representation

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

ax.grid(alpha=0.4)
ax.set(xlim=(-25, 25), ylim=(-25, 25))


# Plot the orthogonal vectors
ax.annotate('', xy=v1, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.annotate('', xy=v2, xytext=(0, 0), arrowprops=dict(facecolor='g'))
plt.show()
```

# Projection
# Part-III

# v2 projection on v1

```python
v1_mag = np.sqrt(np.sum(v1*v1))
p_v1 = np.dot(v1 , v2)*v1/v1_mag
p_v1
```

# v1 projection on v2

```python
v2 = np.array(v2)
v2_mag = np.sqrt(np.sum(v2*v2))
p_v2 = np.dot(v2 , v1)*v2/v2_mag
p_v2
```

# Orthogonal Projection

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

ax.grid(alpha=0.4)
ax.set(xlim=(-8, 8), ylim=(-8, 8))


# Plot the orthogonal vectors
ax.annotate('', xy=v1, xytext=(0, 0), arrowprops=dict(facecolor='r'))
ax.annotate('', xy=v2, xytext=(0, 0), arrowprops=dict(facecolor='g'))
ax.annotate('', xy=p_v1, xytext=(0, 0), arrowprops=dict(facecolor='y'))
ax.annotate('', xy=p_v2, xytext=(0, 0), arrowprops=dict(facecolor='b'))

plt.show()
```

# Part-V

```python
v3 = np.random.randint(3,7,(1,2))[0]
v4 = np.random.randint(3,7,(1,2))[0]
print(v3)
print(v4)
```

```python
v3_mag = np.sqrt(np.sum(v3*v3))
p_v3 = np.dot(v3 , v4)*v3/v3_mag
p_v3
```
